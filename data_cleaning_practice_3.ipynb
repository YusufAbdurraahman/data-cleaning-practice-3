Data cleaning

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df[df["gender"] == "Other"]
```

```python
df = df.drop(df[df["gender"] == "Other"].index)
```

```python
df[df["gender"] == "Other"]
```

```python
df = df.drop(columns = "id")
df.head(2)
```

```python
df.columns = df.columns.str.replace(" ","", regex=False).str.lower()
```

```python
df.head(2)
```

```python
df = df.rename(columns = {"heartdisease":"heart_disease","evermarried" :"ever_married","worktype": "work_type","Residencetype" : "Residence_type", "avgglucoselevel" : "avg_glucose_level","smokingstatus" : "smoking_status"
})
df.head(1)
```

```python
df.columns = df.columns.str.capitalize()
df.head(2)
```

```python
df.columns = df.columns.str.replace("_"," ").str.title().str.replace(" ","_")
df.head(2)
```
